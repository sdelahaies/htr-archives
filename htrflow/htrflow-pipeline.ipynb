```python
from htrflow.pipeline.pipeline import Pipeline
from htrflow.volume.volume import Collection
import yaml

collection = Collection(["ordonnances-malte-01.jpg"])

# Define pipeline configuration
config = yaml.safe_load("""
steps:
- step: Segmentation
  settings:
    model: yolo
    model_settings:
      model: sylvain471/yolov9-line-seg-typewriter
- step: TextRecognition
  settings:
    model: TrOCR
    model_settings:
      model: sylvain471/trocr-typewriter-5
- step: Export
  settings:
    format: json
    dest: outputs
""")



# Process with pipeline
pipe = Pipeline.from_config(config)
collection = pipe.run(collection)

print(collection)
```
